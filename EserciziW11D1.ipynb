**Esercizio** 

Carichiamo in un DataFrame il dataset **iris.csv** dei *beginner_datasets* e calcoliamo: la media della lunghezza dei petali di tutto il dataset; la media della lunghezza dei petali per ogni specie di Iris, utilizzando il metodo **.groupby()**; media, minimo e massimo della larghezza dei sepali per ogni specie, utilizzando **.groupby()** e **.agg()**.

```python
df_iris = pd.read_csv(r'C:\Users\Utente\Desktop\Epicode\Python\Datasets\beginner_datasets\iris.csv')
df_iris
```

```python
pet_len_setosa = df_iris['species'] == 'Iris-setosa'
pet_len_versicolor = df_iris['species'] == 'Iris-versicolor'
pet_len_virginica = df_iris['species'] == 'Iris-virginica'
```

```python
# Senza GROUPBY
print('Media lunghezza petali iris setosa:',round(df_iris.loc[pet_len_setosa, 'petal_length'].mean(), 2))
print('Media lunghezza petali iris versicolor:',round(df_iris.loc[pet_len_versicolor, 'petal_length'].mean(), 2))
print('Media lunghezza petali iris virginica:',round(df_iris.loc[pet_len_virginica, 'petal_length'].mean(), 2))
```

```python
media_setosa = df_iris.loc[pet_len_setosa, 'petal_length'].mean()
media_versicolor = df_iris.loc[pet_len_versicolor, 'petal_length'].mean()
media_verginica = df_iris.loc[pet_len_virginica, 'petal_length'].mean()
lungh_pet_dataset_media = (media_setosa + media_versicolor + media_verginica)/3   # calcolo la media alla vecchia maniera eheh
print('Lunghezza media petali del dataset:', round(lungh_pet_dataset_media, 2))
```

```python
# Con metodo GROUPBY
pet_len = df_iris.groupby('species')['petal_length'].mean()
# Calcolo la media della colonna selezionata tra parentesi quadre raggruppando per specie
```

```python
round(pet_len, 2) # arrotondo il risultato ottenuto
```

```python
df_iris.groupby('species')['sepal_width'].agg(['min', 'max', 'mean']).round(2)
# Minimo, massimo e media della larghezza dei sepali per ciascuna tipologia di iris
```

**CRIM**: tasso di criminalità pro capite per città;

**RAD**: indice di accessibilità alle autostrade radiali;

**TAX**: aliquota dell'imposta fondiaria sul valore pieno per 10.000 dollari; 

**PTRATIO**: rapporto alunni-insegnanti per città;

**BLACK**: la percentuale di neri per città; 

**LSTAT**: % di popolazione di condizione più bassa;

**MEDV**: valore mediano delle case, espresso in migliaia di dollari.

```python
dep = departments_df.set_index('department_id')
empl = employees_df.set_index('department_id')
# Il metodo join unisce le tabelle tramite gli indici, per tale motivo prima di unirle dobbiamo prima trasformare la colonna in comune 
# negli indici di ciasscuna tabella (set_index).
```

```python
join_dep = dep.join(other = empl, on = 'department_id')
```

```python
join_dep
```

Le differenze tra join e merge non le vediamo in output ma nel procedimento: nel join, che unisce tabelle mediante gli indici, dobbiamo fare in modo che la colonna in comune su cui unire sia l'indice di entrambe le tabelle.
In questo caso in output non avremo una tabella con indice *position-based* ma *label-based* (la colonna in comune).
